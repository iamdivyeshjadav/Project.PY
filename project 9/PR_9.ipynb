```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class SalesDataAnalyzer:
    def __init__(self):
        self.data = None
        self.current_figure = None
        self.FILE_NAME = "Q1_Sales_Data.csv"

    def __del__(self):
        plt.close('all')

    def load_data(self, file_path=None):
        target_file = file_path if file_path else self.FILE_NAME
        try:
            self.data = pd.read_csv(target_file)
            print(f"\nDataset '{target_file}' Loaded Successfully!\n")
        except Exception as e:
            print(f"\nError Loading File: {e}\n")

    def explore_data(self):
        if self.data is None:
            print("\nPlease Load The Dataset First.\n")
            return
            
        while True:
            print("\nExplore Data")
            print("1. The First 5 Rows")
            print("2. The Last 5 Rows")
            print("3. Column Names")
            print("4. Data Types")
            print("5. Basic Info Summary")
            print("0. Exit")
            
            try:
                sub_choice = int(input("\nEnter Choice: "))
            except ValueError:
                print("\nInvalid Input. Please Enter A Number.\n")
                continue
                
            match sub_choice:
                case 1:
                    print(f"\n{self.data.head()}\n")
                case 2:
                    print(f"\n{self.data.tail()}\n")
                case 3:
                    print(f"\n{self.data.columns.tolist()}\n")
                case 4:
                    print(f"\n{self.data.dtypes}\n")
                case 5:
                    print("\n")
                    self.data.info()
                    print("\n")
                case 0: 
                    break
                case _:
                    print("\nInvalid Choice.\n")

    def clean_data(self):
        if self.data is None:
            print("\nPlease Load The Dataset First.\n")
            return
            
        while True:
            print("\nHandle Missing Data")
            print("1. Total Missing Values")
            print("2. Fill Missing Values With Mean")
            print("3. Drop Rows With Missing Values")
            print("4. Replace Missing Values With Specific Value")
            print("0. Exit")
            
            try:
                sub_choice = int(input("\nEnter Choice: "))
            except ValueError:
                print("\nInvalid Input. Please Enter A Number.\n")
                continue
                
            match sub_choice:
                case 1:
                    missing = self.data.isna().sum()
                    if missing.sum() == 0:
                        print("\nNo missing values found in the dataset!\n")
                    else:
                        print(f"\nMissing Values Per Column:\n{missing}\n")
                case 2:
                    num_cols = self.data.select_dtypes(include=[np.number]).columns
                    self.data[num_cols] = self.data[num_cols].fillna(self.data[num_cols].mean())
                    print("\nMissing Numeric Fields Filled With Column Averages.\n")
                case 3:
                    self.data.dropna(inplace=True)
                    print("\nRows With Missing Values Dropped.\n")
                case 4:
                    val = input("Enter the value to replace missing data: ")
                    self.data.fillna(val, inplace=True)
                    print(f"\nMissing values replaced with {val}.\n")
                case 0: 
                    break
                case _:
                    print("\nInvalid Choice.\n")

    def dataframe_operations(self):
        if self.data is None:
            print("\nPlease Load The Dataset First.\n")
            return

        while True:
            print("\nDataFrame Operations")
            print("1. Convert Column to Numpy Array & Demonstrate Slicing")
            print("2. Mathematical Operations (Scale Sales by Factor)")
            print("3. Combine Current Data with Dummy DataFrame")
            print("4. Split Data by Region")
            print("5. Search, Sort, & Filter Data")
            print("6. Apply Aggregate Functions (sum, mean, count)")
            print("7. Create Pivot Table Summary")
            print("0. Exit")

            try:
                sub_choice = int(input("\nEnter Choice: "))
            except ValueError:
                print("\nInvalid Input. Please Enter A Number.\n")
                continue

            if sub_choice == 0:
                break

            match sub_choice:
                case 1:
                    col = input("Enter numeric column name (e.g., Sales): ")
                    if col in self.data.columns:
                        arr = self.data[col].to_numpy()
                        print(f"\nConverted Array (First 5 elements): {arr[:5]}")
                        print(f"Slicing Example (Elements 2 to 5): {arr[1:5]}\n")
                    else:
                        print("\nColumn not found.\n")
                case 2:
                    col = input("Enter numeric column to perform mathematical scaling: ")
                    if col in self.data.columns:
                        try:
                            factor = float(input("Enter scaling factor (e.g., 1.1 for 10% increase): "))
                            self.data[col] = self.data[col] * factor
                            print(f"\nColumn '{col}' successfully multiplied by element-wise factor {factor}.\n")
                        except ValueError:
                            print("\nInvalid factor.\n")
                    else:
                        print("\nColumn not found.\n")
                case 3:
                    dummy_data = pd.DataFrame({
                        'SalesID': [999], 'Product': ['Product Z'], 'Region': ['Global'], 'Sales': [1000], 'Year': [2026]
                    })
                    self.data = pd.concat([self.data, dummy_data], ignore_index=True)
                    print("\nSuccessfully combined current data with another DataFrame using concat.\n")
                case 4:
                    col = input("Enter column name to split by (e.g., Region): ")
                    if col in self.data.columns:
                        unique_vals = self.data[col].unique()
                        print(f"\nSplitting data into {len(unique_vals)} structural data subsets based on '{col}':")
                        for val in unique_vals:
                            subset = self.data[self.data[col] == val]
                            print(f"Subset [{val}] contains {len(subset)} records.")
                        print("\n")
                    else:
                        print("\nColumn not found.\n")
                case 5:
                    col = input("Enter column to search/filter by: ")
                    if col in self.data.columns:
                        val = input(f"Enter exact value to filter for in '{col}': ")
                        filtered = self.data[self.data[col].astype(str) == val]
                        print(f"\nFiltered Results:\n{filtered}\n")
                        sort_col = input("Enter column name to sort entire dataset by: ")
                        if sort_col in self.data.columns:
                            self.data.sort_values(by=sort_col, inplace=True)
                            print(f"\nDataset sorted by '{sort_col}' successfully.\n")
                    else:
                        print("\nColumn not found.\n")
                case 6:
                    col = input("Enter column to aggregate (e.g., Sales): ")
                    if col in self.data.columns:
                        try:
                            print(f"\nAggregated Results for '{col}':")
                            print(f"Sum  : {self.data[col].sum()}")
                            print(f"Mean : {self.data[col].mean()}")
                            print(f"Count: {self.data[col].count()}\n")
                        except Exception as e:
                            print(f"\nAggregation failed: {e}\n")
                case 7:
                    index_col = input("Enter Row Index column for Pivot Table (e.g., Region): ")
                    val_col = input("Enter Numerical Values column (e.g., Sales): ")
                    if index_col in self.data.columns and val_col in self.data.columns:
                        pivot = self.data.pivot_table(values=val_col, index=index_col, aggfunc=['sum', 'mean'])
                        print(f"\nGenerated Pivot Table:\n{pivot}\n")
                    else:
                        print("\nInvalid Columns Entered.\n")

    def mathematical_operations(self):
        pass

    def statistical_analysis(self):
        if self.data is None:
            print("\nPlease Load The Dataset First.\n")
            return
        print("\nDescriptive Statistics")
        print(self.data.describe())
        num_cols = self.data.select_dtypes(include=[np.number]).columns
        if len(num_cols) > 0:
            print("\nAdvanced Statistical Computations:")
            print(f"Standard Deviation:\n{self.data[num_cols].std()}")
            print(f"Variance:\n{self.data[num_cols].var()}")
            print(f"Quantiles (0.25, 0.5, 0.75):\n{self.data[num_cols].quantile([0.25, 0.5, 0.75])}")
        print("\n")

    def visualize_data(self):
        if self.data is None:
            print("\nPlease Load The Dataset First.\n")
            return
            
        while True:
            print("\nData Visualization")
            print("1. Bar Plot")
            print("2. Line Plot")
            print("3. Scatter Plot")
            print("4. Pie Chart")
            print("5. Histogram")
            print("6. Stack Plot")
            print("7. Seaborn Enhanced Visualizations (Heatmap/Boxplot)")
            print("0. Exit")
            
            try:
                sub_choice = int(input("\nEnter Choice: "))
            except ValueError:
                print("\nInvalid Input. Please Enter A Number.\n")
                continue
                
            if sub_choice == 0:
                break
                
            plt.close('all')
            sns.set_theme(style="darkgrid")
            
            try:
                match sub_choice:
                    case 1:
                        x = input("Enter x-axis column name: ")
                        y = input("Enter y-axis column name: ")
                        if x in self.data.columns and y in self.data.columns:
                            self.current_figure, ax = plt.subplots(figsize=(8, 5))
                            sns.barplot(data=self.data, x=x, y=y, ax=ax, palette='Blues_r')
                            ax.set_title(f"Bar Plot: {y} by {x}")
                            ax.set_xlabel(x)
                            ax.set_ylabel(y)
                        else:
                            print("\nInvalid Columns Entered.\n")
                            continue
                            
                    case 2:
                        x = input("Enter x-axis column name: ")
                        y = input("Enter y-axis column name: ")
                        if x in self.data.columns and y in self.data.columns:
                            self.current_figure, ax = plt.subplots(figsize=(8, 5))
                            sns.lineplot(data=self.data, x=x, y=y, marker='o', color='g', ax=ax)
                            ax.set_title(f"Line Plot: {y} over {x}")
                            ax.set_xlabel(x)
                            ax.set_ylabel(y)
                        else:
                            print("\nInvalid Columns Entered.\n")
                            continue
                            
                    case 3:
                        x = input("Enter x-axis column name: ")
                        y = input("Enter y-axis column name: ")
                        if x in self.data.columns and y in self.data.columns:
                            self.current_figure, ax = plt.subplots(figsize=(8, 5))
                            sns.scatterplot(data=self.data, x=x, y=y, ax=ax, color='purple')
                            ax.set_title(f"Scatter Plot: {x} vs {y}")
                            ax.set_xlabel(x)
                            ax.set_ylabel(y)
                            print("\nGenerating scatter plot...")
                            print("Scatter plot displayed successfully!")
                        else:
                            print("\nInvalid Columns Entered.\n")
                            continue
                            
                    case 4:
                        cat_col = input("Enter Categorical column: ")
                        num_col = input("Enter Numerical column: ")
                        if cat_col in self.data.columns and num_col in self.data.columns:
                            summary = self.data.groupby(cat_col)[num_col].sum()
                            self.current_figure, ax = plt.subplots(figsize=(6, 6))
                            ax.pie(summary, labels=summary.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
                            ax.set_title(f"Pie Chart: Total {num_col} Share")
                        else:
                            print("\nInvalid Columns Entered.\n")
                            continue

                    case 5:
                        num_col = input("Enter numeric column name for Histogram: ")
                        if num_col in self.data.columns:
                            self.current_figure, ax = plt.subplots(figsize=(8, 5))
                            sns.histplot(data=self.data, x=num_col, kde=True, ax=ax, color='orange')
                            ax.set_title(f"Histogram: Distribution of {num_col}")
                        else:
                            print("\nColumn not found.\n")
                            continue

                    case 6:
                        x_col = input("Enter x-axis baseline numeric or temporal column: ")
                        y_col = input("Enter y-axis stacking numeric column: ")
                        if x_col in self.data.columns and y_col in self.data.columns:
                            sorted_df = self.data.sort_values(by=x_col)
                            self.current_figure, ax = plt.subplots(figsize=(8, 5))
                            ax.stackplot(sorted_df[x_col], sorted_df[y_col], labels=[y_col], colors=['cyan'])
                            ax.set_title(f"Stack Plot Chart: {y_col} Stacked over {x_col}")
                            ax.legend(loc='upper left')
                        else:
                            print("\nInvalid Columns Entered.\n")
                            continue

                    case 7:
                        print("\nEnhanced Visualizations\n1. Box Plot\n2. Correlation Heatmap")
                        sub = input("Select sub-option: ")
                        self.current_figure, ax = plt.subplots(figsize=(8, 5))
                        if sub == '1':
                            x_b = input("Enter categorical column: ")
                            y_b = input("Enter numeric column: ")
                            if x_b in self.data.columns and y_b in self.data.columns:
                                sns.boxplot(data=self.data, x=x_b, y=y_b, ax=ax, palette='Set2')
                                ax.set_title(f"Box Plot: {y_b} distribution by {x_b}")
                        elif sub == '2':
                            num_data = self.data.select_dtypes(include=[np.number])
                            if not num_data.empty:
                                sns.heatmap(num_data.corr(), annot=True, cmap='coolwarm', ax=ax)
                                ax.set_title("Correlation Heatmap")
                            else:
                                print("\nNo numeric data available for heatmap.\n")
                                continue
                        else:
                            print("\nInvalid sub-option.\n")
                            continue

                plt.tight_layout()
                plt.show()
                
            except Exception as e:
                print(f"\nPlotting Error: {e}. Please check your inputs.\n")

    def save_visualization(self):
        if self.current_figure is None:
            print("\nNo Visualization Found! Please Generate A Plot First.\n")
            return
            
        print("\nSave Visualization")
        filename = input("Enter file name to save the plot (e.g., scatter_plot.png): ")
        try:
            self.current_figure.savefig(filename, bbox_inches='tight')
            print(f"Visualization saved as {filename} successfully!\n")
        except Exception as e:
            print(f"\nFailed To Save File: {e}\n")

def main():
    analyzer = SalesDataAnalyzer()
    
    while True:
        print("\n========== Data Analysis & Visualization Program ==========")
        print("Please select an option:")
        print("1. Load Dataset")
        print("2. Explore Data")
        print("3. Perform DataFrame Operations")
        print("4. Handle Missing Data")
        print("5. Generate Descriptive Statistics")
        print("6. Data Visualization")
        print("7. Save Visualization")
        print("8. Exit")
        print("===========================================================")
        
        choice = input("\nEnter your choice: ")
        
        match choice:
            case '1':
                path = input("Enter the path of the dataset (CSV file): ")
                analyzer.load_data(path)
            case '2':
                analyzer.explore_data()
            case '3':
                analyzer.dataframe_operations()
            case '4':
                analyzer.clean_data()
            case '5':
                analyzer.statistical_analysis()
            case '6':
                analyzer.visualize_data()
            case '7':
                analyzer.save_visualization()
            case '8':
                print("\nExiting the program. Goodbye!\n")
                break
            case _:
                print("\nInvalid Choice, Please Try Again.\n")

if __name__ == "__main__":
    main()
```
